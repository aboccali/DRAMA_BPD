```python
from neuroHarmonize import harmonizationLearn
from neuroHarmonize import harmonizationApply, loadHarmonizationModel
import neuroHarmonize as nh
import pandas as pd
import numpy as np
```

```python
my_data  = pd.read_csv('IMAGING_RR_QC_LIGHT_VERSION_for_Suicide_Squad_per_armonizzazione_TOT.csv', index_col='Sub-id')
columns_name = my_data.columns
my_data.replace(['.', 'nan', 'Nan', ' '], np.nan, inplace = True)
my_data = my_data.astype(float)
my_data = np.array(my_data)
covars = pd.read_csv('covariates_TOT.csv', index_col='Sub-id')
```

```python
#Inizializza un DataFrame vuoto per i risultati armonizzati
harmonized_results = pd.DataFrame(index=range(my_data.shape[0]), columns=range(my_data.shape[1]))

#Loop attraverso le colonne di my_data
for col_idx in range(my_data.shape[1]):
    #Seleziona la colonna corrente di my_data
    col_data = my_data[:, col_idx]
    
    #Trova le righe senza NaN nella colonna corrente di my_data
    valid_rows = ~np.isnan(col_data)
    
    #Filtra my_data e il DataFrame delle covariates per le righe valide
    data_valid = col_data[valid_rows].reshape(-1, 1)
    covars_valid = covars[valid_rows]
    
    #Applica l'armonizzazione ai dati validi
    model_col_idx, data_adj = harmonizationLearn(data_valid, covars_valid, eb=True, ref_batch='site_A', return_s_data=False)
    nh.saveHarmonizationModel(model_col_idx, f'MY_MODEL_{columns_name[col_idx]}')
    #Aggiorna harmonized_results con i valori armonizzati
    harmonized_results.loc[valid_rows, col_idx] = data_adj.flatten()

#Visualizza i risultati
print(harmonized_results)
harmonized_results.to_excel(f'harmonized_results_TOT_PROVA.xlsx')
```

```python
my_data_1subj  = pd.read_csv('IMAGING_RR_QC_LIGHT_VERSION_for_Suicide_Squad_per_armonizzazione_TOT_only_1_SUBJ.csv', index_col='Sub-id')
my_data_1subj.replace(['.', 'nan', 'Nan', ' '], np.nan, inplace = True)
my_data_1subj = my_data_1subj.astype(float)
my_data_1subj = np.array(my_data_1subj)
covars_1subj = pd.read_csv('covariates_TOT_PROVA_only_1_SUBJ.csv', index_col='Sub-id')
```

```python
#Inizializza un DataFrame vuoto per i risultati armonizzati
harmonized_results = pd.DataFrame(index=range(my_data_1subj.shape[0]), columns=range(my_data_1subj.shape[1]))

#Loop attraverso le colonne di my_data_1subj
for col_idx in range(my_data_1subj.shape[1]):
    #Seleziona la colonna corrente di my_data_1subj
    col_data = my_data_1subj[:, col_idx]
    
    #Trova le righe senza NaN nella colonna corrente di my_data_1subj
    valid_rows = ~np.isnan(col_data)
    
    #Filtra my_data_1subj e il DataFrame delle covariates per le righe valide
    data_valid = col_data[valid_rows].reshape(-1, 1)
    covars_valid = covars_1subj[valid_rows]
    print(f"Column {col_idx}: data_valid shape: {data_valid.shape}, covars_valid shape: {covars_valid.shape}")
    #Applica l'armonizzazione ai dati validi
    my_model_col_idx = loadHarmonizationModel(f'MY_MODEL_{columns_name[col_idx]}')
    data_harmonized_1_subj = harmonizationApply(data_valid, covars_valid, my_model_col_idx)
    
    #Aggiorna harmonized_results con i valori armonizzati
    harmonized_results.loc[valid_rows, col_idx] = data_harmonized_1_subj.flatten()

#Visualizza i risultati
print(harmonized_results)
harmonized_results.to_excel(f'harmonized_results_1_subj_PROVA.xlsx') 
```
